```python
# 1️⃣ Movie Ratings (MPAA)
# Rating	Ordinal Value
# G	1 (Most family-friendly)
# PG	2
# PG-13	3
# R	4

# NC-17	5 (Most restricted)
# 2️⃣ TV Ratings (TV Parental Guidelines)
# Rating	Ordinal Value
# TV-Y	1 (For all kids, youngest audience)
# TV-Y7	2
# TV-Y7-FV	3 (Includes fantasy violence)
# TV-G	4
# TV-PG	5
# TV-14	6
# TV-MA	7 (Most restricted, mature audience only)

# 3️⃣ Other Ratings
# Rating	Ordinal Value
# NR (Not Rated)	NaN (No clear restriction level)
# UR (Unrated)	NaN
# #
```

```python
rating_order = {
    'G': 1, 'PG': 2, 'PG-13': 3, 'R': 4, 'NC-17': 5,
    'TV-Y': 1, 'TV-Y7': 2, 'TV-Y7-FV': 3, 'TV-G': 4,
    'TV-PG': 5, 'TV-14': 6, 'TV-MA': 7,
    'NR': None, 'UR': None
}
```

```python
df['Rank_Rating'] = df['Rating'].map(rating_order)
```
